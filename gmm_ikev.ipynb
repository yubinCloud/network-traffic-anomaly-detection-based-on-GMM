```python
import numpy as np
from scipy.stats import norm
from sklearn.mixture import GaussianMixture

data = np.genfromtxt('ikev2-rtt.csv', delimiter=',')
eps = 2.4

X0 = data[:, 0]
X1 = data[:, 1]
X2 = data[:, 2:5]
X3 = data[:, 5]

X0 = X0.reshape(-1, 1)
X1 = X1.reshape(-1, 1)
X2 = np.sum(X2, axis = 1)/3
X2 = X2.reshape(-1, 1)
X3 = X3.reshape(-1, 1)

gm0 = GaussianMixture(n_components = 15, random_state = 0).fit(X0)
mu0 = gm0.means_
sigma0 = gm0.covariances_

gm1 = GaussianMixture(n_components = 2, random_state = 0).fit(X1)
mu1 = gm1.means_
sigma1 = gm1.covariances_

gm2 = GaussianMixture(n_components = 2, random_state = 0).fit(X2)
mu2 = gm2.means_
sigma2 = gm2.covariances_

gm3 = GaussianMixture(n_components = 2, random_state = 0).fit(X3)
mu3 = gm3.means_
sigma3 = gm3.covariances_

gms = [gm0, gm1, gm2, gm3]
mus = [mu0, mu1, mu2, mu3]
sigmas = [sigma0, sigma1, sigma2, sigma3]
# count = 0
# count0 = 0
# count1 = 0
# count2 = 0
# count3 = 0
counts = [0,0,0,0]
ratio = [0,0,0,0]

for i in range(len(data)):
    #每一行数据
    td_ = data[i]
    td0 = td_[0]
    td1 = td_[1]
    td2 = (td_[2] + td_[3] + td_[4])/3
    td3 = td_[5]
    td = [td0, td1, td2, td3]
    probs = []
    for j in range(4):
        #一行的四个数据
        prob_ = gms[j].predict_proba(td[j].reshape(-1, 1))
        #一个数据的不同参数
        sum = 0
        label = gms[j].predict(td[j].reshape(-1, 1))
        if gms[j].n_components>1:
            for k in range(gms[j].n_components):
                sum = sum+norm.pdf(td[j], mus[j][k], sigmas[j][k])
            prob = norm.pdf(td[j], mus[j][label], sigmas[j][label])/sum
        else:
            prob = norm.pdf(td[j], mus[j][label], sigmas[j][label])
        probs.append(prob)
    # prod = 0
    #print('probs:', probs)
    # for pro in probs:
    #     prod = prod+pro
    # if prod>=eps:
    #     count = count+1
    for l in range(4):
        if probs[l]>0.95:
            counts[l] = counts[l]+1
            
    for l in range(4):
        ratio[l] = counts[l]/len(data)

# print('count:', count)
print('counts:',counts)
print('ratiro: ', ratio)
```
